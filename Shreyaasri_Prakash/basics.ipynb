# Basic Usage


### Ollama CLI

#### https://github.com/ollama/ollama

#### The main Ollama command-line interface 
1. ollama - Opens the interactive menu in the terminal.
2. ollama run model - Starts an interactive chat with a model, for example ollama run gemma3.
3. ollama pull model - Downloads a model.
4. ollama rm model - Removes a downloaded model.
5. ollama ls - Lists downloaded models.
6. ollama ps - Lists models currently loaded/running in memory.
7. ollama stop model - Stops a running model.
8. ollama serve - Starts the Ollama server.
9. ollama launch - Launches integrations interactively, such as ollama launch codex or ollama launch claude.
10. ollama show model -  Inspect model parameters 
11. ollama cp model new-model name -  renaming the model
