## 3. Decoding Parameters

### Role-Based Inference with Temperature Adjustment
 Task 1: Modify the temperature setting to see how it affects the creativity of the model's responses.

 How Temperature Works: ranges from 0.0 to 2.0
 0.0 – 0.3 → More accurate, deterministic answers
 0.5 – 0.7 → Balanced, creative but still reliable
 0.8 – 1.2 → More creative, varied, sometimes less factual
 1.3+ → Highly creative, may produce unexpected or less relevant answers

 Task 2: Experiment with different max_tokens values to control the length of the model's responses.

 ranges depends on the model’s context window size and the specific use case.
 Common settings include:
 50-100 tokens → Short answers or summaries
 150-300 tokens → Detailed responses or explanations
 400+ tokens → In-depth content or extended conversations

 Task 3: Adjust top_p and top_k settings to explore their impact on response diversity and relevance.

 A) Top_p: Controls sampling diversity (nucleus sampling).
 'top_p': 0.9 -> means the model considers the smallest set of words whose cumulative probability is at least 90%.
 Lower values make the output more focused and deterministic, while higher values increase diversity.

 B) Top_k: Limits the number of highest probability vocabulary tokens to consider.
 'top_k': 40 -> means the model only considers the top 40 tokens at each step.
 Higher values allow for more diversity, while Lower = safer, more deterministic.

```python
import ollama

response = ollama.chat(
    model='llama3.2:latest',
    messages=[
        # Setting behaviour of the model
        {
            'role': 'system',
            'content': (
                'You are a knowledgeable travel guide who provides concise, '
                'accurate, and friendly recommendations.'
            )
        },

        # User prompt 
        {'role': 'user', 'content': 'What are the top five tourist attractions in Berlin?'}
    ],
    options={
        'temperature': 0.2,   # Adjust temperature here
        'max_tokens': 100,
        'top_p': 0.9,
        'top_k': 40
    }
)

print(response['message']['content'])
```

## 4. Interactive Multi-Turn Chat Loop 

### Project Goal
 When you run this code, the model acts as a Berlin-only travel guide. It will answer your questions about Berlin’s attractions, but if you ask about any other city, it will politely decline and redirect you back to Berlin. The chat continues interactively, remembering previous messages.

Task:1  If you notice the model still answering questions about other cities, you may need to make the system prompt more strict or more explicit so the model stays within the Berlin-only domain

Task:2  Test the chat loop by asking questions about Berlin as well as other cities to ensure the model adheres to the specified behavior

Task:3  Add another city to the system prompt where the model can also provide recommendations, but still restricts all other cities except for the ones that are mentioned


```python
import ollama

messages = [
    {
        "role": "system",
        "content": (
            "You are a knowledgeable travel guide who provides concise, "
            "accurate, and friendly recommendations only about Berlin and nothing else."
        )
    }
]

while True:
    user_input = input("You: ")

    # User message
    messages.append({"role": "user", "content": user_input})

    # Response from model
    response = ollama.chat(model="your-model_name", messages=messages)

    answer = response["message"]["content"]
    print("Assistant:", answer)
```
